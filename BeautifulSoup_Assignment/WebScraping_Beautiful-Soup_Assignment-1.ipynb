## 1) Write a python program to display all the header tags from wikipedia.org.

```python
url='https://en.wikipedia.org/wiki/Main_Page'
page = requests.get(url)
```

```python
# page content
soup = BeautifulSoup(page.content)
soup
```

```python
# Scrap the Header tag
header=[]
for i in soup.find_all('span',class_="mw-headline"):  # we can also write like this
    header.append(i.text)                             # header=[i.text for i in soup.find_all(class_="mw-headline")]

header
```

```python
# Create DataFrame
df=pd.DataFrame(header,columns=["Heading of Wiki"])
df
```

## 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

b) Top 10 ODI Batsmen along with the records of their team and rating.

c) Top 10 ODI bowlers along with the records of their team and rating.

## a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page = requests.get(url)
# see content in page5
soup = BeautifulSoup(page.content)
#scrape team names
team = soup.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] 
points = [] 
ratings = []
new_list = []

# first place team number of matches
for i in soup.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)
    
# first place team points
for i in soup.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
    
# first place team ratings
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))

# other teams number of matches and points
for i in soup.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
    
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) # other teams matches
    points.append(new_list[i+1]) # other teams points

    # other teams ratings
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
    
# Make data frame of top 10 ICC teams
icc=pd.DataFrame({"Team_Name": team_name, "Match":matches, "Points":points, "Ratings":ratings})
icc    
```

### 7) Write a python program to scrape mentioned news details from
https://www.cnbc.com/world/?region=world :

i) Headline

ii) Time

iii) News Link

```python
# Send url request and download page source code
News_soup=BeautifulSoup(requests.get("https://www.cnbc.com/world/?region=world").content,"html.parser")
# Scrap Headlines
Headline=[i.text for i in News_soup.find_all(class_="LatestNews-headline") ] 
Time=[i.text for i in News_soup.find_all("span",class_="LatestNews-wrapper") ] # Scrap Time
# Scrap news Links
news = News_soup.find_all('div', {'class','LatestNews-container'})
lnk = [each.a.get('href') for each in news]
# Create data frame
news1=pd.DataFrame({'Headlines':Headline,'Time':Time,"Link":lnk})
news1
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :

i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
#send get request to the webpage server to get the source code of the page and download the page content
soup=BeautifulSoup(requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special').content,'html.parser')
soup
```

```python
# Scrap Restaurant name
name=[i.text.split(',')[0] for i in soup.find_all('div',class_='restnt-info cursor')] 
name
```

```python
# Scrap location
location=[i.text for i in soup.find_all('div',class_='restnt-loc ellipsis')]  
location
```

```python
# scrap url of image
image=[i['data-src'] for i in soup.find_all("img", class_="no-img")] 
image
```

```python
# scrap cuisine
cuisine=[i.text.split('|')[1] for i in soup.find_all('span',class_="double-line-ellipsis")] 
cuisine
```

```python
# scrap rating
rating=[i.text for i in soup.find_all('div',class_='restnt-rating rating-4')] 
rating
```

```python
df=pd.DataFrame({'Restaurant Name':name,'Cuisine':cuisine,'Location':location,"Rating":rating,"Image_url": image})
df
```
